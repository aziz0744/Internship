```python
#Ans 11
def factorial_iterative(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

# Example usage
number = int(input("Enter a number: "))
print(f"The factorial of {number} is {factorial_iterative(number)}")
```

```python
# Ans 12 
def is_prime(n):
    # Edge cases
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    
    # Check from 5 to sqrt(n)
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    
    return True

def main():
    number = int(input("Enter a number: "))
    
    if number <= 0:
        print("Please enter a positive integer.")
    elif is_prime(number):
        print(f"{number} is a prime number.")
    else:
        print(f"{number} is a composite number.")

if __name__ == "__main__":
    main()
```

```python
# Ans 13
import re

def is_palindrome(s):
    # Normalize the string
    normalized_string = re.sub(r'[^a-zA-Z0-9]', '', s).lower()
    
    # Check if the normalized string is equal to its reverse
    return normalized_string == normalized_string[::-1]

def main():
    input_string = input("Enter a string: ")
    
    if is_palindrome(input_string):
        print(f'"{input_string}" is a palindrome.')
    else:
        print(f'"{input_string}" is not a palindrome.')

if __name__ == "__main__":
    main()
```

```python
# Ans 14
import math

def find_third_side(a, b, hypotenuse=False):
    if hypotenuse:
        # Calculate the hypotenuse when two sides are given
        return math.sqrt(a**2 + b**2)
    else:
        # Calculate the other leg when one leg and the hypotenuse are given
        if a >= b:
            return math.sqrt(a**2 - b**2)
        else:
            return math.sqrt(b**2 - a**2)

def main():
    print("Enter the sides of the right-angled triangle:")
    side1 = float(input("Enter the length of the first side: "))
    side2 = float(input("Enter the length of the second side: "))
    
    # Determine if the provided sides include the hypotenuse
    choice = input("Are the provided sides the two legs of the triangle (type 'legs') or one leg and the hypotenuse (type 'hypotenuse')? ").strip().lower()
    
    if choice == 'legs':
        result = find_third_side(side1, side2, hypotenuse=True)
        print(f"The length of the hypotenuse is: {result:.2f}")
    elif choice == 'hypotenuse':
        result = find_third_side(side1, side2, hypotenuse=False)
        print(f"The length of the other leg is: {result:.2f}")
    else:
        print("Invalid input. Please specify 'legs' or 'hypotenuse'.")

if __name__ == "__main__":
    main()

    
```

```python
# Ans 15
def count_characters(string):
    # Create a dictionary to hold the frequency of each character
    frequency = {}
    
    # Iterate through each character in the string
    for char in string:
        if char in frequency:
            frequency[char] += 1
        else:
            frequency[char] = 1
    
    return frequency

def main():
    # Get input from the user
    input_string = input("Enter a string: ")
    
    # Count character frequencies
    freq = count_characters(input_string)
    
    # Print the frequencies
    print("Character frequencies:")
    for char, count in freq.items():
        print(f"'{char}': {count}")

if __name__ == "__main__":
    main()
```
